```python
import pandas as pd

UNSW_NB15_PATH = '/home/sgeraci/Desktop/datasets/UNSW-NB15/raw/UNSW-NB15_{pcap_file_no}.csv'
UNSW_NB15_COLUMN_NAMES = '/home/sgeraci/Desktop/datasets/UNSW-NB15/UNSW-NB15_features.csv'
```

```python
# lighter version
columns_names = pd.read_csv(UNSW_NB15_COLUMN_NAMES, header=None)
columns_names = columns_names[1].iloc[1:].tolist()
flows = pd.Series()
for pcap_file_no in range(1, 5):
    print('')
    print(f'Loading pcap file numer {pcap_file_no}: {UNSW_NB15_PATH.format(pcap_file_no=pcap_file_no)}... ', end='')
    pcap = pd.read_csv(UNSW_NB15_PATH.format(pcap_file_no=pcap_file_no), low_memory=False)
    pcap.columns = columns_names
    pcap = pcap[['srcip', 'sport', 'dstip', 'dsport', 'proto', 'synack', 'ackdat']]

    pcap['sport']  = pd.to_numeric(pcap['sport'], errors='coerce').astype('Int64')
    pcap['dsport'] = pd.to_numeric(pcap['dsport'], errors='coerce').astype('Int64')
    pcap['proto']  = pcap['proto'].str.lower()
    print(f'Loaded.')

    tcp_handshaked = pcap.loc[
                            (pcap['synack']!=0)&
                            (pcap['ackdat']!=0)
                            ]
    if(flows.empty and tcp_handshaked.size > 0):
        print(f'src_to_dst_flow has been set to:\n {tcp_handshaked.iloc[:1]}. End.\n')
        flows = tcp_handshaked.iloc[:10]
        break;
```
